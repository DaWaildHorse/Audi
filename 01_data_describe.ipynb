```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa import arima
import seaborn as sns
```

```python
df = pd.read_csv('C:/CODE/Python/Py_hackathon/data/Absenteeism_at_work.csv')
```

```python
df.info()
```

```python
grouped_month = df.groupby('Month of absence').size().reset_index(name='Count')

plt.figure(figsize=(10,6))
sns.lineplot(data=grouped_month, x='Month of absence', y='Count', marker='o')
plt.title('Serie de Tiempo de Conteos por Mes de Ausencia')
plt.xlabel('Mes de Ausencia')
plt.ylabel('Conteo')
plt.show()
```

```python
grouped_day = df.groupby('Day of the week').size().reset_index(name='Count')

plt.figure(figsize=(10,6))
sns.lineplot(data=grouped_day, x='Day of the week', y='Count', marker='o')
plt.title('Serie de Tiempo de Conteos por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Conteo')
plt.show()
```

```python
grouped_season = df.groupby('Seasons').size().reset_index(name='Count')

plt.figure(figsize=(10,6))
sns.lineplot(data=grouped_season, x='Seasons', y='Count', marker='o')
plt.title('Serie de Tiempo de Conteos por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Conteo')
plt.show()
```

```python
grouped_season = df.groupby('Age').size().reset_index(name='Count')

plt.figure(figsize=(10,6))
sns.lineplot(data=grouped_season, x='Age', y='Count', marker='o')
plt.title('Serie de Tiempo de Conteos por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Conteo')
plt.show()
```

```python
grouped_season = df.groupby('Service time').size().reset_index(name='Count')

plt.figure(figsize=(10,6))
sns.lineplot(data=grouped_season, x='Service time', y='Count', marker='o')
plt.title('Serie de Tiempo de Conteos por años de servicio')
plt.xlabel('Temporada')
plt.ylabel('Conteo')
plt.show()
```

```python
corr_matrix = df.corr()
corr_matrix_filtered = corr_matrix.mask(np.abs(corr_matrix) < 0.5)


plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix_filtered, cmap='viridis', annot=True, mask=corr_matrix_filtered.isnull(), cbar=True)

# Mostrar el gráfico
plt.title('Heatmap de Correlaciones con Valor Absoluto Mayor a 0.5')
plt.show()
```
